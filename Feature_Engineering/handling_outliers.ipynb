## 5 number Summary and Box plot

```python
import numpy as np
```

```python
marks = [45, 32, 56, 75, 89, 54, 32, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74]
minimum, Q1, median, Q3, maximum = np.quantile(marks, [0, 0.25, 0.50, 0.75, 1])
print(minimum, Q1, median, Q3, maximum)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_fence = Q1-1.5*(IQR)
higher_fence = Q3+1.5*(IQR)
print(lower_fence, '<------->' , higher_fence)
```

There are no values in the list which are outside the lower and higher fence range

```python
import seaborn as sns
sns.boxplot(marks)
```
